# the series in stationary

```python
from statsmodels.tsa.seasonal import seasonal_decompose
mul_result=seasonal_decompose(df.iloc[0:1000,:],model='multiplicative',period=1)
add_result=seasonal_decompose(df.iloc[0:1000,:],model='addictive',period=1)
mul_result
```

```python
from statsmodels.tsa.stattools import kpss
kpss_test=kpss(df.iloc[0:1000,:])
print('Test Statistics',kpss_test[0])
print('P value',kpss_test[1])
```
